```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the labels
y_encoded = label_encoder.fit_transform(df['labels'])

# Check the mapping of original labels to encoded labels
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label Mapping:", label_mapping)

# Replace the original labels with encoded labels in the DataFrame
df['labels2'] = y_encoded
```

```python

# Assuming df is your DataFrame containing lip features and labels
# Splitting the data into features and labels
from sklearn.model_selection import train_test_split
X = df.drop(columns=['frames', 'labels','person','labels2'])  # Assuming 'frames' is not a feature
y = df['labels']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python

import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score


# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing data
X_test_scaled = scaler.transform(X_test)
```

# XGBoost

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Initialize XGBoost classifier
xgb_classifier = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)

# Train the classifier
xgb_classifier.fit(X_train_scaled, y_train)

# Make predictions
y_pred = xgb_classifier.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("XGBoost Accuracy:", accuracy)
```

# CatBoost

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score

# Initializing CatBoost classifier
catboost_classifier = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=6)

catboost_classifier.fit(X_train_scaled, y_train, verbose=100)

# Making predictions
catboost_preds = catboost_classifier.predict(X_test_scaled)

# Evaluating the classifier
catboost_accuracy = accuracy_score(y_test, catboost_preds)

print("CatBoost Accuracy:", catboost_accuracy)
```

# Light Boost

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import accuracy_score

# Create dataset for LightGBM
train_data = lgb.Dataset(X_train_scaled, label=y_train)
test_data = lgb.Dataset(X_test_scaled, label=y_test)

# Set parameters for LightGBM
params = {
    'objective': 'multiclass',  # Multiclass classification
    'num_class': len(set(y_train)),  # Number of classes
}

# Train LightGBM model
num_round = 100  # Number of boosting rounds
lgb_model = lgb.train(params, train_data, num_round)

# Predictions
y_pred = lgb_model.predict(X_test_scaled)
y_pred_class = [int(x.argmax()) for x in y_pred]

# Accuracy
accuracy = accuracy_score(y_test, y_pred_class)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import accuracy_score

# Assuming X_train_scaled, X_test_scaled, y_train, y_test are already defined

# Create dataset for LightGBM
train_data = lgb.Dataset(X_train_scaled, label=y_train)
test_data = lgb.Dataset(X_test_scaled, label=y_test)

# Set parameters for LightGBM
params = {
    'objective': 'multiclass',  # Multiclass classification
    'num_class': len(set(y_train)),  # Number of classes
    'metric': 'multi_logloss',  # Metric for evaluation
    'boosting_type': 'gbdt',  # Gradient Boosting Decision Tree
    'num_leaves': 31,  # Maximum number of leaves in one tree
    'learning_rate': 0.05,  # Learning rate
    'feature_fraction': 0.9,  # Feature fraction
    'bagging_fraction': 0.8,  # Bagging fraction
    'bagging_freq': 5,  # Frequency for bagging
    'verbose': 0  # Verbose output
}

# Train LightGBM model
num_round = 100  # Number of boosting rounds
lgb_model = lgb.train(params, train_data, num_round, valid_sets=[test_data], early_stopping_rounds=10)

# Predictions
y_pred = lgb_model.predict(X_test_scaled, num_iteration=lgb_model.best_iteration)
y_pred_class = [int(x.argmax()) for x in y_pred]

# Accuracy
accuracy = accuracy_score(y_test, y_pred_class)
print("Accuracy:", accuracy)
```
