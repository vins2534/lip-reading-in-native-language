```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Assuming you have X_train_scaled, X_test_scaled, y_train, and y_test defined somewhere

# Define the parameter grid to search
param_grid_svc = {
    'C': [0.1, 1, 10],             # Regularization parameter
    'kernel': ['linear', 'rbf','sigmoid','poly'],   # Kernel type
    'gamma': ['scale', 'auto']     # Kernel coefficient for 'rbf' kernel
}

# Initialize SVC model
svc_gs = SVC()

# Initialize GridSearchCV
grid_search_svc = GridSearchCV(estimator=svc_gs, param_grid=param_grid_svc, cv=5, n_jobs=-1, scoring='accuracy')

# Perform grid search to find the best parameters
grid_search_svc.fit(X_train_scaled, y_train)

# Get the best parameters and best estimator
best_params_svc = grid_search_svc.best_params_
best_estimator_svc = grid_search_svc.best_estimator_

# Make predictions using the best estimator
best_preds_svc = best_estimator_svc.predict(X_test_scaled)

# Calculate accuracy using the best estimator
best_accuracy_svc = accuracy_score(y_test, best_preds_svc)

print("Best Parameters:", best_params_svc)
print("Best Accuracy:", best_accuracy_svc)
```
